> # Genotype Dataframe Without Encoding

# Group_1

```python
### Extract Genotype Information for each and every single SNP in every single patient
### and store these Informations from each patient in a list to then concatenate them in a single Dataframe 

G1_list= []
for file in tqdm(os.listdir(Gp1_merged)):
    file_path = os.path.join(Gp1_merged,file)
    Sample_ID = os.path.basename(file).split('.')[0]

## Read every single file in a separate df
## print df.shape to get idea about ur df dimensions.
    df =pd.read_csv(file_path) 
    print(Sample_ID, df.shape)                         
    
    ## Before apply ordinal encoding which is additive encoding counting number of minor alleles i.e. 
    ## genotype "0/0" will be coded as '0' caz it contains zero minor alelles
    ## genotype "0/1" will be coded as '1' caz it contains 1 minor alelles
    ## genotype "1/1" will be coded as '2' caz it contains 2 minor alelles
    ## first replace different genotypes as following:
    print ('GT: ')
    print(df['GT'].value_counts())
    df['GT'] = df['GT'].str.replace('0/2','0/1')
    df['GT'] = df['GT'].str.replace('1/2','1/1') 
    df['GT'] = df['GT'].str.replace('0|1','0/1', regex=False) 
    df['GT'] = df['GT'].str.replace('1|1','1/1', regex=False)
    df['GT'] = df['GT'].str.replace('1|2','1/1', regex=False) 
    
## combine CHROM + POS in one variable
## Create new column called SNP_ID and insert it into df
    
    Head = df["CHROM"] + '-' + df["POS"].astype(str)   
    df.insert(loc=1, column= 'SNP_ID', value= Head) 
    
## Now use multiple filters on ur df to include only SNPs that has     
## 'FILTER' == PASS and that has 
## minor allele frequency 'AF' >= 0.01  .....???
## 'QUAL' >= 30.0
## Also exclude SNPs with missing genotypes like '1/.' or '0/.' 0r './.' 
## i.e include onle those SNPs with genotype '0/1' or '1/1'
## include SNPs related to specific chromosomes that is known to be associated with our disease of interest 
##'CHROM' == ['chr1', 'chr2', 'chr3', 'chr9', 'chr22', 'chrX']

    df = df[(df['FILTER']=='PASS')&(df['QUAL']>=30)& ((df['GT'] == '0/1') | (df['GT'] == '1/1'))] 
   
    chromosome = ['chr1', 'chr2', 'chr3', 'chr9', 'chr22', 'chrX']
    df = df[df["CHROM"].isin(chromosome)]
    
    print ('----------------')
    print ('GT: ')
    print(df['GT'].value_counts())
    
    print(Sample_ID, df.shape)          ## Note that the df.shape now has fewer rows for SNPs that was filtered out
    df = df[['SNP_ID','GT']]
    df = df.rename(columns = {'GT': Sample_ID})
    df = df.T  
    df.columns= df.iloc[0]
    df = df.drop(['SNP_ID'])
    print(Sample_ID, df.shape)
    print('*********************************************************')
    G1_list.append(df)

```

```python
%%time
## Dataframe containing all Genotypes of All SNPs in all samples

G1_raw_genotype = pd.concat(G1_list,axis=0)
G1_raw_genotype
```

> ## Replace nan values in Genotype DataFrame

```python
## from my point of view the null values could be replaced by the genotype '0/0'
## as an example to explain the SNP = chr1-10002921	exists only in sample 'G2_Sample_42_SNPs' and NAN in all others
## since it does not appear in others this means that their genotype at thie exact SNP are homozygous wildtype '0/0'
## and thus were not called by the variant caller in those samples so we replase them by '0/0'

G1_G2_G3_NoNull = G1_G2_G3_Sorted.fillna('0/0')
G1_G2_G3_NoNull
```
